```python
#Team: Sneha Susan Varghese & Chinmay Bake 

import requests
import urllib.request
from bs4 import BeautifulSoup 
import pandas as pd
#importing required set of libraries for scraping datd from the web
```

```python
url3 = "https://en.wikipedia.org/wiki/List_of_rugby_union_playing_countries" #define the url/link from which data is to be scraped 
r1 = requests.get(url3) #getting a response from the page
```

```python
list1=list()  #declare an empty list 
soup2 = BeautifulSoup(r1.content, "html.parser") #source all the HTML content of the page
tbl = soup2.find_all('table')[0]   #print all the content in the HTML with 'table' notation in it with its index specified
rw = tbl.find_all('tr') #in the above declared variables, we could lookout for 'tr' notation or essentially data inside the table rows


for tr in rw:
    td = tr.find_all('td')      
    row = [i.text for i in td]
    list1.append(row)    
    
#now we extract data from each cell in the row; by searching 'td' notation in the previously declared 'rw' variable. 
#every text in the 'td' variable with table data is sequeantially appended in the empty list which we have declared previously
```

```python
df = pd.DataFrame(list1,columns=['Country','No. of Clubs','Registered Players','Referees','Under 13 Boys','Under 13 Girls','Teen Males','Teen Females','Senior Males','Senior Females','Total Males','Total Females','Total Players'])
#define a pandas dataframe constituting of the list which has been created. Declare the column headers in it. 
df.set_index('Country',inplace=True)
#set index as the 'Country' column in the dataframe
df.to_csv('Sneha_Chinmay_Assignment3(3).csv')
#Import the dataframe to a csv file

#NOTE: After opening the CSV file, Wrap Text button is default enabled for data cells starting from second column. So the data won't be displayed upfront. 
       #Kindly click on any of the datacells, select all the data rows & columns and disable the wrap text option. Thank you!. 
    
```
